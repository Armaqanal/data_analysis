```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
try:
  df=pd.read_csv("diabetes(1).csv")
  # print(df.isnull().sum())
  df_clean=df.copy()
  numeric_cols=df_clean.select_dtypes(include=['int64','float64']).columns
  df_clean[numeric_cols]=df_clean[numeric_cols].fillna(df_clean[numeric_cols].mean())
  # print(df_clean.isnull().sum())
except FileNotFoundError as e:
   print(e)
```

```python
df_clean.head(5)
```

```python
df1=df_clean[["Outcome", "Insulin", "Age", "Glucose", "BMI", "BloodPressure"]]
df1
```

```python
# correlation matrix

df_correlation=df1.corr()
print(df1)
print(df_correlation.columns)
```

```python
# creat heatmap
df1=sns.heatmap(df_correlation, annot=True,cbar_kws={'label':'orrelation Coefficient'},center=0)
select_outcome = df_correlation['Outcome']

select_outcome = select_outcome.drop('Outcome')
max_feature = select_outcome.idxmax()
max_value = select_outcome.max()

print("max feature", max_feature)
print("amount:", max_value)

```
