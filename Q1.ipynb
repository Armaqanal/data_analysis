```python
import numpy as np
import random

cities = ["Tehran", "Mashhad", "Isfahan", "Tabriz", "Shiraz", "Ahvaz"]
days = 30
daily_temperature = np.random.randint(15, 41, size=(days, len(cities)))
humidity = np.random.randint(20, 81, size=(days, len(cities)))
rainfall = np.random.randint(0, 51, size=(days, len(cities)))

# print("Daily Temperature:\n", daily_temperature)
# print("\nHumidity:\n", humidity)
print("\nRainfall:\n", rainfall[0])


```

```python
import pandas as pd
daily=np.arange(1,31)
pd_obj=[]
for i,city in enumerate(cities):
  for day in daily:
    pd_obj.append({'city':city,
                   'days':day,
                   'daily_temperature':daily_temperature[day-1][i],
                   'humidity':humidity[day-1][i],
                   'rainfall':rainfall[day-1][i]})

df=pd.DataFrame(pd_obj)
print(df.head(15))
```

```python
mean_temp_by_city = df.groupby('city')['daily_temperature'].mean()
mean_humidity_by_city = df.groupby('city')['humidity'].mean()
mean_rain_by_city = df.groupby('city')['rainfall'].mean()
print(mean_temp_by_city)
```

```python
hottest_city_amount = mean_temp_by_city.max()
hottest_city_name= mean_temp_by_city.idxmax()

coldest_city_amount = mean_temp_by_city.min()
coldest_city_name= mean_temp_by_city.idxmin()

# print("hottest city name :" ,hottest_city_name, "\nhottest city amount: ",hottest_city_amount)
# print("coldest city name :" ,coldest_city_name, "\ncoldest city amount: ",coldest_city_amount)
```
