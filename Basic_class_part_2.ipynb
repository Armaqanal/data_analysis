```python
import pandas as pd
from scipy.spatial.distance import euclidean
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris_dataset.csv")

all_features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)","target"]
df_q6=pd.DataFrame(df,columns=all_features)
print(df_q6.head())

p_test=df.iloc[-1,:-1]
training_features=df.iloc[:-1,:-1]
true_class = df.iloc[:-1, -1]

distances = [euclidean(p_test, training_features.iloc[i])
             for i in range(len(training_features))]

min_distance_three = sorted(enumerate(distances), key=lambda x: x[1])[:3]
three_index = [i for i, d in min_distance_three]
nearest_neighbors = training_features.iloc[three_index]
predicted_class = true_class.iloc[three_index]

print("Before normalization:")
print(nearest_neighbors, predicted_class.values)
```

```python
for column in training_features:
    df[column] = (df[column] - df[column].mean()) / df[column].std()
    # print(df[column])

p_test_norm = df.iloc[-1, :-1]
training_norm = df.iloc[:-1, :-1]

distances_norm = [euclidean(p_test_norm, training_norm.iloc[i])
                  for i in range(len(training_norm))]

min_distance_three_norm = sorted(enumerate(distances_norm), key=lambda x: x[1])[:3]
three_index_norm = [i for i, d in min_distance_three_norm]
nearest_neighbors_norm = training_norm.iloc[three_index_norm]
predicted_class_norm = true_class.iloc[three_index_norm]

print("After normalization:")
print(nearest_neighbors_norm, predicted_class_norm.values)
```
