```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np




x=pd.DataFrame(iris.data.features)
y=pd.DataFrame(iris.data.targets)

df=pd.concat([x,y],axis=1)
df.head(5)
```

```python
missing_value=df.isna().sum(axis=0)
missing_value
```

```python
corr_pearson_pental=df['petal length'].corr(df['petal width'],method='pearson').round(3)
corr_pearson_sepal=df['sepal length'].corr(df['sepal width'],method='pearson').round(3)
print(corr_pearson_pental)
print(corr_pearson_sepal)

```

```python
df["petal_ratio"]=df["petal length"] / df["petal width"]
```

HEATMAP


```python
df_updated=df.drop(["class","petal_ratio"],axis=1)
correlation_matrix=df_updated.corr()
data_mask=np.triu(np.ones_like(correlation_matrix,dtype=bool))
# print(data_mask)

sns.heatmap(correlation_matrix,vmin = -1, vmax = 1, center = 0,mask=data_mask,annot=True)
plt.show()
```

PAIRPLOT

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
sns.pairplot(
    df[features],
    diag_kind='kde',
    plot_kws={'color': '#6A5ACD'})

# plt.figure(figsize=(5, 5))
```

BOXPLOT

```python
sns.boxplot(df,x='class',y="petal_ratio")
```
